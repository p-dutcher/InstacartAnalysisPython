```python
# use group by to:
# 1. Split data into groups based on criteria
# 2. Apply a function tp each group separately
# 3. Combine results into dataframe or alternative data structure (or new column in current df)
```

```python
# aggregating data w agg()

df.groupby('department_id').agg({'order_number':['mean']})
```

```python
#  the above return is the average # of orders per user for each department id
#  can help us determine best selling departments
```

```python
# conducting the above without using agg()

df.groupby('department_id')['order_number'].mean()
```

```python
# performing multiple aggregations - simply add more arguments to code

df.groupby('department_id').agg({'order_number':['mean','min','max']})
```

```python
# Aggregating Data with tranform()
# creating a flag column (allows for segmented analysis)

# criteria for loyalty flag:

# max orders over 40 = "loyal customer"
# max orders over 10, but less than or equal to 40 = "regular customer"
# max orders under or equal to 10 = "new customer"
```

```python
# Steps:
# 1. split data into groups based on user_id column
# 2. apply transform() on the order_number column to generate the max orders for each user
# 3. create a new "max_order" column to merge results of aggregation

#can be done in a single line of code:

df['max_order'] = df.groupby(['user_id'])['order_number'].transform(np.max)

# first create new column in data frame
# then groupby desired column
# finally transform on the column being aggregated                                                                                             
```

```python
df.head(15)
```

```python
# deriving columns with loc()

df.loc[df['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df.loc[(df['max_order'] <= 40) & (df['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df.loc[df['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df['loyalty_flag'].value_counts(dropna = False)
```

```python
# displaying only certain columns:

df[['user_id','loyalty_flag','order_number']].head(60)
```

```python
# create a loyalty flag for existing customers using transform() and loc()

# criteria for loyalty flag:

# max orders over 40 = "loyal customer"
# max orders over 10, but less than or equal to 40 = "regular customer"
# max orders under or equal to 10 = "new customer"
```

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# deriving columns with loc()

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# create a new spending flag: for each user, based on average spend amount, label high or low spender

#Flag criterias:
# mean of product prices purchased by a user is <10 then 'Low spender'
# mean of product prices purchased by a user is => 10 then 'High spender'
```

```python
# first create a new column for average spent by each user

ords_prods_merge['mean_prod_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(25)
```

```python
#Flag criterias:
# mean of product prices purchased by a user is <10 then 'Low spender'
# mean of product prices purchased by a user is => 10 then 'High spender'

ords_prods_merge.loc[ords_prods_merge['mean_prod_price'] < 10, 'spender_type'] = 'Low Spender'
ords_prods_merge.loc[ords_prods_merge['mean_prod_price'] >= 10, 'spender_type'] = 'High Spender'
```

```python
ords_prods_merge['spender_type'].value_counts(dropna = False)
```

```python
# the majority of spenders are low spenders (average product purchase price is under $10)
```

```python
ords_prods_merge[['user_id','loyalty_flag','mean_prod_price','spender_type']]
```

```python
# create an order frequency flag
# this tracks order frequency using median of days_since_since_prior_order

#criteria:
# if median is >20 then labeled "non-frequent customer"
# if median is >10 or <= 20 then "regular customer"
# if median is <= 10, then "frequent customer"
```

```python
# will create a new column: "median_since_last_order"

ords_prods_merge['median_since_last_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# create flag & new column using loc()

# new column w flag: "customer_frequency_type"

ords_prods_merge.loc[ords_prods_merge['median_since_last_order'] > 20, 'customer_frequency_type'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_since_last_order'] <= 20) & (ords_prods_merge['median_since_last_order'] > 10), 'customer_frequency_type'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_since_last_order'] <= 10, 'customer_frequency_type'] = 'Frequent customer'
```

```python
# checking counts of new column:
ords_prods_merge['customer_frequency_type'].value_counts(dropna = False)
```

```python
# checking new column:

ords_prods_merge[['user_id','median_since_last_order', 'customer_frequency_type']].head(60)
```
